## NYC public school test results - data analysis

This data contains SAT scores of each school in NYC.

```python
# import libraries
import pandas as pd
```

```python
# import data
schools = pd.read_csv("schools.csv")
schools.head()
```

```python
# Which NYC schools have the best math results? (with at least 80% of the max score, which is 800)
bms = schools[schools['average_math'] >= 0.8 * 800]
best_math_schools = bms[['school_name', 'average_math']].sort_values('average_math', ascending=False)
best_math_schools
```

```python
# What are the top 10 performing schools based on the combined SAT scores?
schools_copy = schools
schools_copy['total_SAT'] = schools_copy['average_math'] + schools_copy['average_reading'] + schools_copy['average_writing']
top_10_schools = schools_copy[['school_name', 'total_SAT']].sort_values('total_SAT', ascending=False).head(10)
top_10_schools
```

```python
# Which single borough has the largest standard deviation in the combined SAT score?
borough = schools_copy[['borough', 'total_SAT']].groupby('borough').std()
top_borough = borough.sort_values('total_SAT', ascending=False).head(1)
top_borough_raw = schools_copy[['borough', 'school_name', 'total_SAT']][schools_copy['borough'] == top_borough.reset_index()['borough'].values[0]] # reset_index to treat 'borough' as a normal column
largest_std_dev = top_borough_raw.groupby('borough').agg({'school_name':'count', 'total_SAT':'mean'})
largest_std_dev['std_SAT'] = round(top_borough['total_SAT'].values[0], 2)
largest_std_dev = largest_std_dev.rename(columns={'school_name':'num_schools', 'total_SAT':'average_SAT'})
largest_std_dev['average_SAT'] = round(largest_std_dev['average_SAT'], 2)
largest_std_dev
```
